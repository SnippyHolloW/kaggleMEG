```python
from scipy.io import loadmat
from pylab import imshow, show
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = 15, 10
import glob
subject_files = glob.glob("data/train*.mat")
print subject_files

d = loadmat(subject_files[0])
print d.keys()
time = numpy.linspace(float(d['tmin']), float(d['tmax']), num=d['X'].shape[2])
# d['X'].shape[2]/float(d['sfreq'][0])
```

```python
# MAKE SURE you have memory before loading all the data!! (next cell)
data = d['X']
labels = d['y'].ravel()
'Aprox Mem needed : %.4f Gb' % (len(subject_files) * data.nbytes / float(2**30))
```

```python
for suj in subject_files:
    d = loadmat(suj)  # get it from https://www.kaggle.com/c/decoding-the-human-brain/download/train_01_06.zip
    data = np.concatenate([data, d['X']], axis=0)
    labels = np.concatenate([labels, d['y'].ravel()], axis=0)
'Used Mem : %.4f Gb' % (data.nbytes / float(2**30))   
```

```python
print data.shape
print labels.shape
```

## apparently a P1... need to improove electrode selection

```python
# Notice how the activity of interest is around 6e-13 (or ~600 fT), order of magnitude of endogenous MEG activity
# More than that is almost centainly noise (of course this is an average, and single trials could go higher,
# but I would put a threshold of 1000 fT for signal...)

m = data.mean(axis=0)
elecmean = m[200:300,:].mean(axis=0)
ax = subplot(111)
xticks_pos = range(0, elecmean.shape[0], 25)
xticks_time = map(lambda x: "%.02f" % (x/d['sfreq'][0][0] - 0.5), range(0, elecmean.shape[0], 25))
plt.xticks(xticks_pos, xticks_time)
ax.plot(elecmean)
```
